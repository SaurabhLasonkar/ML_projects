```python
#Density Distribution of Numerical features for Fraud and valid transaction
feature_to_plot  = ['V1','V2','V3'] #choose feature to plot

for feature in feature_to_plot:
    plt.figure(figsize=(6,8))
    sns.kdeplot(df[df['Class']==0][feature],label='Valid',shade=True)
    sns.kdeplot(df[df['Class']==1][feature],label='Fraud',shade=True)
    plt.title(f'Density distribution of {feature} by class')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```
